```python
import pandas as pd
```

```python
# take an instance of the xl file having data of movies
xl_file = 'movies.xls'
#reading data from the first sheet setting 0th column as the index column
data = pd.read_excel(xl_file,sheet_name=0,index_col=0)
# reading top few entries of data
#data.head()
```

```python
# reading 2nd and 3rd sheets respectively
data2 = pd.read_excel(xl_file,sheet_name=1,index_col=0)
data3 = pd.read_excel(xl_file,sheet_name=2,index_col=0)
#data2.head()
#data3.head()
```

```python
#concating the data of each sheet into 1
movies = pd.concat([data,data2,data3])
type(movies)
```

```python
movies.head()
# to get total no of rows and columns of movies
movies.shape
# to get values from the dataframes
movies_data = movies.get_values()
type(movies_data)
print(movies_data[0])
type(movies_data[0])
print(movies_data[0][0])
```

```python
# doing the same task using ExcelFile class by parsing the sheets
xls_file = pd.ExcelFile(xl_file)
movies_sheets = []
# parse all the individual sheets in xls_file and append data in empty list above
for sheets in xls_file.sheet_names:
    print(sheets)
    movies_sheets.append(xls_file.parse(sheets))
    
print(type(movies_sheets))
#print(movies_sheets[0])
```

```python
# to sort values as per a specific column
sort_by_gross = movies.sort_values(['Gross Earnings'],ascending=False)
sort_by_gross['Gross Earnings'].head(10)
```
